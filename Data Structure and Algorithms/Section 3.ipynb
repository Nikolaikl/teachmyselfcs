```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# R31
X = np.linspace(1,50) # for 10 Tasks
Y1 = 8 * X
Y2 = 4 * X* np.log(X)
Y3 = 2 * X** 2
Y4 = X ** 3
Y5 = 2 ** X

plt.plot(np.log(X),np.log(Y1))
plt.plot(np.log(X),np.log(Y2))
plt.plot(np.log(X),np.log(Y3))
plt.plot(np.log(X),np.log(Y4))
plt.plot(np.log(X),np.log(Y5))
plt.legend(['8n', '$4nlogn$', '$2n^2$','$n^3$','$2^n$'])
plt.title('Asymptotic Runtime Analysis - Log Log Scaled')
plt.show()
```
